```python
import numpy as np
import cv2

from matplotlib import pyplot as plt

%matplotlib inline
```

```python
frame = np.load('../images/sample-1.npy')
```

```python
gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
gray = gray[0:150, 0:300]
plt.imshow(cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)[1], cmap='gray')
```

```python
c = cv2.findContours(gray, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
len(c)
```

```python
a = cv2.moments(c[0])
```

```python
a['m10'] / a['m00']
```

```python
a['m01'] / a['m00']
```
